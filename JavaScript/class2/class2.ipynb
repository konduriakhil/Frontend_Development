## Where to Add JavaScript in an HTML Document?

Use the `<script></script>` tag to add JavaScript in HTML.

### Placement Options

- In the `<head>` tag
- In the `<body>` tag
- In **both**

> üìù **If you want the script to run before the body is loaded**, place the `<script>` tag inside the `<head>` section.

> üìù **If you want the script to run after the entire body is loaded**, place the `<script>` tag just **before the closing `</body>` tag**.

## Printing Output

- Use `document.write("text")` to print directly to the web page.

```html
<!DOCTYPE html>
<html>
<head>
  <title>Welcome to JavaScript</title>
  <script>
    // JavaScript code before body loads
    document.write("Hello");
  </script>
</head>
<body>
  <h1>Welcome to JavaScript</h1>

  <script>
    // JavaScript code after body content
    document.write('<h1>Welcome to Js, from Sunadh</h1>');
    console.log("This is Ram");
  </script>
</body>
</html>
